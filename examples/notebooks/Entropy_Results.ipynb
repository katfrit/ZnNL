```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    '10': {'entropy': [1.4299815, 1.0649279], 'model': {'accuracy': 0.43800002336502075, 'loss': 2.549513816833496}}, 
    '20': {'entropy': [1.6265097, 1.2852912], 'model': {'accuracy': 0.527999997138977, 'loss': 2.0837535858154297}}, 
    '30': {'entropy': [1.7681414, 1.2330437], 'model': {'accuracy': 0.5740000009536743, 'loss': 2.213181972503662}}, 
    '50': {'entropy': [1.9276254, 1.2277381], 'model': {'accuracy': 0.7350000143051147, 'loss': 2.4009249210357666}}, 
    '100': {'entropy': [2.2563639, 1.3267068], 'model': {'accuracy': 0.8010000586509705, 'loss': 1.0252071619033813}}}
```

### Accuracy vs $\Delta S$

```python
ds = []
acc = []

for value in data.values():
    ds.append(value['entropy'][0] - value['entropy'][1])
    acc.append(value['model']['accuracy'])
    
plt.plot(ds, acc, 'o')
plt.xlabel("$\Delta S$")
plt.ylabel("Accuracy")
plt.show()
```

### $S_{0}$ vs Dataset Size

```python
s0 = []
size = []

for item in data:
    s0.append(data[item]['entropy'][0])
    size.append(item)
    
plt.plot(size, s0, 'o')
plt.xlabel("DS Size")
plt.ylabel("$S_{0}$")
plt.show()
```
